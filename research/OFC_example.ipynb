```python
import numpy as np
import SOC.models
```

```python
modele={}
#alpha=[0.1,0.2,0.25]
alpha=[0.1,0.2]
for m in alpha:
    modele[m] = SOC.models.OFC(50, 1, m)
    modele[m].run(300000)
```
