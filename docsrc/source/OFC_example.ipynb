# OFC model

```python
import numpy as np
import SOC.models
```

```python
modele={}
#alpha=[0.1,0.2,0.25]
alpha=[0.1,0.2]
for m in alpha:
    modele[m] = SOC.models.OFC(1, m, L=50)
    modele[m].run(300000)
```

```python
modele[0.1].get_exponent(hist_num=50, smooth_width = 25, plot = True)['exponent']
```

```python
modele[0.1].get_exponent(col='NumberOfReleases', hist_num=50, smooth_width = 25, plot = True)['exponent']
```
